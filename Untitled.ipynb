```python
import requests, io, pandas as pd, xml.etree.ElementTree as ET
from datetime import datetime

# 1) Your list (paste/replace if you add more)
YOUR_ROLE_IDS = [
'RO101','RO102','RO103','RO104','RO105','RO106','RO107','RO108','RO109','RO11','RO110','RO111','RO114','RO116',
'RO117','RO119','RO12','RO122','RO123','RO126','RO128','RO131','RO132','RO134','RO136','RO137','RO138','RO140',
'RO141','RO142','RO144','RO146','RO147','RO148','RO149','RO15','RO150','RO153','RO154','RO155','RO157','RO158',
'RO159','RO161','RO162','RO166','RO167','RO168','RO169','RO171','RO172','RO173','RO175','RO176','RO177','RO179',
'RO180','RO181','RO182','RO185','RO189','RO190','RO191','RO197','RO198','RO200','RO209','RO21','RO210','RO211',
'RO212','RO213','RO214','RO215','RO216','RO217','RO218','RO22','RO221','RO222','RO223','RO227','RO228','RO229',
'RO230','RO231','RO232','RO233','RO234','RO235','RO236','RO24','RO246','RO247','RO248','RO249','RO25','RO250',
'RO251','RO252','RO255','RO256','RO257','RO258','RO259','RO260','RO261','RO262','RO266','RO267','RO268','RO269',
'RO270','RO272','RO273','RO274','RO275','RO276','RO277','RO279','RO280','RO281','RO282','RO283','RO284','RO285',
'RO286','RO287','RO288','RO289','RO29','RO290','RO291','RO292','RO293','RO294','RO296','RO297','RO298','RO299',
'RO30','RO300','RO301','RO302','RO303','RO304','RO305','RO306','RO307','RO308','RO309','RO31','RO310','RO311',
'RO312','RO313','RO314','RO315','RO316','RO317','RO318','RO319','RO320','RO321','RO322','RO323','RO324','RO325',
'RO326','RO327','RO328','RO33','RO34','RO35','RO36','RO37','RO38','RO39','RO40','RO57','RO59','RO60','RO65','RO67',
'RO7','RO71','RO72','RO76','RO80','RO82','RO83','RO87','RO88','RO89','RO90','RO91','RO92','RO93','RO94','RO96',
'RO98','RO99'
]

# 2) Fetch the authoritative roles list (ORD roles CodeSystem)
ROLES_XML_URL = "https://directory.spineservices.nhs.uk/ORD/2-0-0/roles?_format=xml"
r = requests.get(ROLES_XML_URL, timeout=60)
r.raise_for_status()

# 3) Parse XML -> rows
root = ET.fromstring(r.content)
rows = []
# The XML looks like: <CodeSystem> <Roles> <Role> <id>RO177</id> <code>177</code> <displayName>...</displayName> <primaryRole>true</primaryRole> </Role> ... 
for role in root.findall(".//Role"):
    rid   = (role.findtext("id") or "").strip()
    code  = (role.findtext("code") or "").strip()
    name  = (role.findtext("displayName") or "").strip()
    prim  = (role.findtext("primaryRole") or "").strip().lower()
    rows.append({
        "RoleId": rid,
        "CodeNumeric": code,
        "RoleName": name,
        "IsPrimaryRole": True if prim == "true" else False
    })

df_all = pd.DataFrame(rows).sort_values(["IsPrimaryRole","RoleId"], ascending=[False, True]).reset_index(drop=True)

# 4) Filter to your list (keep only those ids you provided)
df_sel = df_all[df_all["RoleId"].isin(YOUR_ROLE_IDS)].copy().reset_index(drop=True)

# 5) Save CSVs
ts = datetime.now().strftime("%Y%m%d")
df_all.to_csv(f"ods_roles_all_{ts}.csv", index=False)
df_sel.to_csv(f"ods_roles_selected_{ts}.csv", index=False)

# 6) Save Excel with 2 sheets
with pd.ExcelWriter(f"ods_roles_{ts}.xlsx", engine="openpyxl") as xl:
    df_all.to_excel(xl, sheet_name="AllRoles", index=False)
    df_sel.to_excel(xl, sheet_name="YourList", index=False)

print("Wrote:",
      f"ods_roles_all_{ts}.csv,",
      f"ods_roles_selected_{ts}.csv,",
      f"ods_roles_{ts}.xlsx")
df_sel.head(10)
```
